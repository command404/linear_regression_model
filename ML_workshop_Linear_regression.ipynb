```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/ML_workshop_csv.csv")
df
```

```python
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color="red",marker = "+")
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["area"]],df.price) #the 2d array is necessary to supply the data frame.on the basis of area were telling it to predict prices(thats why we wrote area first)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_ #this shoes the value of M in y = Mx + b
```

```python
reg.intercept_ #this shows the value of b in y = Mx+b
```

```python
#y=M*x+b sub M and b, x is the value we pass in reg.predict
135.78767123*3300+180616.4383561643
#the answer we get is the value of y
```

```python
plt.xlabel("area", fontsize = 20)
plt.ylabel("price", fontsize = 20)
plt.scatter(df.area,df.price,color = "red",marker = "+")
plt.plot(df.area,reg.predict(df[["area"]]),color = "blue")
```
